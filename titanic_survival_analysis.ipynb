# 🚢 Titanic Survival Analysis — Descriptive Statistics

An exploratory analysis of the Titanic passenger dataset, surfacing the main factors associated with passenger survival.

## Objective

Produce a descriptive statistical analysis of the data, highlighting the most relevant patterns associated with passenger survival.

## Libraries

Importing the core libraries for statistical data analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Loading the dataset — Titanic

The dataset contains real data on the Titanic passengers.

```python
df = sns.load_dataset("titanic")
df
```

## Exploratory Data Analysis

The analysis below highlights the most relevant insights found in the data, each illustrated with a chart: overall survival, survival rate by sex, survival by passenger class, fare distribution by class, family aboard, and survival by port of embarkation.

```python
sobreviventes = df.survived.value_counts()
sobreviventes.index = ['Did not survive', 'Survived']
sobreviventes.plot(kind="pie", autopct='%.1f%%', colors=['green', 'gray'])
plt.title('Survival')
plt.show()
```

```python
homens = df[df['sex'] == 'male']
total_homens = len(homens)
contagem_homens = homens.survived.value_counts()
porcentagem_homens = (contagem_homens / total_homens) * 100
porcentagem_homens.index = [f"Died | {porcentagem_homens[0]:.1f}%", f"Survived | {porcentagem_homens[1]:.1f}%"]
porcentagem_homens.plot(kind='bar', color=['gray', 'green'], rot=0)
plt.grid(axis='y', linestyle=':')
plt.yticks(range(0, 80, 15))
plt.title('Survival Rate — Men')
plt.show()
```

```python
mulheres = df[df['sex'] == 'female']
total_mulheres = len(mulheres)
contagem_mulheres = mulheres.survived.value_counts()
porcentagem_mulheres = (contagem_mulheres / total_mulheres) * 100
porcentagem_mulheres.index = [f"Survived | {porcentagem_mulheres[1]:.1f}%", f"Died | {porcentagem_mulheres[0]:.1f}%"]
porcentagem_mulheres.plot(kind='bar', color=['green', 'gray'], rot=0)
plt.grid(axis='y', linestyle=':')
plt.yticks(range(0, 80, 15))
plt.title('Survival Rate — Women')
plt.show()
```

```python
fclass = df[df['pclass'] == 1]
sclass = df[df['pclass'] == 2]
tclass = df[df['pclass'] == 3]
tclass
```

```python
# first class
total_fclass = len(fclass)
contagem_fclass = fclass.survived.value_counts()
porcentagem_fclass = (contagem_fclass / total_fclass) * 100
```

```python
# second class
total_sclass = len(sclass)
contagem_sclass = sclass.survived.value_counts()
porcentagem_sclass = (contagem_sclass / total_sclass) * 100
```

```python
# third class
total_tclass = len(tclass)
contagem_tclass = tclass.survived.value_counts()
porcentagem_tclass = (contagem_tclass / total_tclass) * 100
```

```python
# Simple colors and horizontal labels
classes_nomes = ['1st Class', '2nd Class', '3rd Class']
taxas_sobrevivencia = [porcentagem_fclass[1], porcentagem_sclass[1], porcentagem_tclass[1]]

# Using predefined colors
plt.bar(classes_nomes, taxas_sobrevivencia, color=['darkgreen', 'green', 'lightgreen'])

# Keep the x labels horizontal
plt.xticks(rotation=0)

plt.grid(axis='y', linestyle=':')
plt.yticks(range(0, 101, 10))
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Percentage (%)')
plt.show()
```

```python
sns.boxplot(data=df, x='pclass', y='fare', showfliers=False, palette='Set2')
plt.title('Fare Distribution by Class (Outliers Removed)')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.xticks([0, 1, 2], ['1st Class', '2nd Class', '3rd Class'])  # Label the classes on the X axis
plt.grid(axis='y', linestyle=':')

plt.show()
```

```python
# Boxplot straight from the family column
df['sibsp'].plot(kind='box', color='green')

# Final touches
plt.title('Distribution of Siblings and Spouses Aboard')
plt.ylabel('Number of People')
plt.grid(axis='y', linestyle=':')

plt.show()
```

```python
# 1. Split the groups by embarkation port
porto_c = df[df['embarked'] == 'C']
porto_s = df[df['embarked'] == 'S']
porto_q = df[df['embarked'] == 'Q']

# 2. Survival percentage for each (index [1] = survived)
taxa_c = (porto_c['survived'].value_counts() / len(porto_c)) * 100
taxa_s = (porto_s['survived'].value_counts() / len(porto_s)) * 100
taxa_q = (porto_q['survived'].value_counts() / len(porto_q)) * 100

# 3. Organize for the chart
portos_nomes = ['Cherbourg (France)', 'Southampton (England)', 'Queenstown (Ireland)']
taxas_finais = [taxa_c[1], taxa_s[1], taxa_q[1]]

# 4. Simple bar chart
plt.bar(portos_nomes, taxas_finais, color=['darkblue', 'blue', 'lightblue'])

plt.xticks(rotation=0)
plt.grid(axis='y', linestyle=':')
plt.yticks(range(0, 101, 10))
plt.title('Survival Rate by Port of Embarkation')
plt.ylabel('Survivors (%)')

plt.show()
```
